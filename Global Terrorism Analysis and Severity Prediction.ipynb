## About the Dataset

Information on more than 180,000 Terrorist Attacks. 

The Global Terrorism Database (GTD) is an open-source database including information on terrorist attacks around the world from 1970 through 2017. The GTD includes systematic data on domestic as well as international terrorist incidents that have occurred during this time period and now includes more than 180,000 attacks. The database is maintained by researchers at the National Consortium for the Study of Terrorism and Responses to Terrorism (START), headquartered at the University of Maryland.

#### Content

Geography: Worldwide

Time period: 1970-2017, except 1993

Unit of analysis: Attack

Variables: >100 variables on location, tactics, perpetrators, targets, and outcomes

Sources: Unclassified media articles (Note: Please interpret changes over time with caution. Global patterns are driven by diverse trends in particular regions, and data collection is influenced by fluctuations in access to media coverage over both time and place.)

Definition of terrorism:

"The threatened or actual use of illegal force and violence by a non-state actor to attain a political, economic, religious, or social goal through fear, coercion, or intimidation."

### 6) Handling missing Values & Data Type Correction

```python
# Fatalities and Injuries: Fill NaN with 0, assuming unreported means no casualties.
df['Fatalities'] = df['Fatalities'].fillna(0)
df['Injuries'] = df['Injuries'].fillna(0)

# City & State/Province: Fill missing with 'Unknown'
df['City'] = df['City'].fillna('Unknown')
df['State/Province'] = df['State/Province'].fillna('Unknown')

# Group Name: Fill missing group names with 'Unknown'
df['Group Name'] = df['Group Name'].fillna('Unknown')

# Target Subtype: Fill missing with 'Unknown' (or 'Not Applicable' if preferred)
df['Target Subtype'] = df['Target Subtype'].fillna('Unknown')

# Summary and Motive: Fill NaN with 'No information provided'
df['Summary'] = df['Summary'].fillna('No information provided')
df['Motive'] = df['Motive'].fillna('No information provided')

# Property Damage: Convert -9 to NaN, then fill NaN with 0 (assuming no reported damage means no damage)
df['Property Damage'] = df['Property Damage'].replace(-9, np.nan).fillna(0).astype(int)

# Doubt_Terrorism: Convert -9 to NaN, then fill with 0 (assuming if not doubted, it's 0)
df['Doubt_Terrorism'] = df['Doubt_Terrorism'].replace(-9, np.nan).fillna(0).astype(int)

# Let's fill with 0 for now as a placeholder, acknowledging this isn't ideal for maps if many are 0.
df['Latitude'] = df['Latitude'].fillna(0)
df['Longitude'] = df['Longitude'].fillna(0)
```

```python
for col in ['Successful Attack', 'Suicide Attack', 'Extended Attack',
            'Individual Attack', 'Vicinity', 'Multiple_Attack',
            'Terrorism_Crit_1', 'Terrorism_Crit_2', 'Terrorism_Crit_3']:
    if col in df.columns: # Check if column exists after initial dropping
        df[col] = df[col].fillna(0).astype(int) # Fill NaN with 0, then convert to int
```

```python
#Convert Year, Month, Day to int. Month/Day can have 0 for unknown, which is fine as int.
df['Year'] = df['Year'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Day'] = df['Day'].astype(int)
```

```python
# Create a new feature combining fatalities and injuries to represent the overall severity of an attack.

df['Total_Casualties'] = df['Fatalities'] + df['Injuries']
print("Created 'Total Casualties' feature.")
```

### 6) Feature Engineering & Preprocessing for Machine Learning

This section prepares our data for various machine learning models. We will:
1.  Define a reduced set of features and target variables.
2.  Handle high-cardinality categorical features (only for the selected few).
3.  Apply One-Hot Encoding.
4.  Standard Scale numerical features.
5.  Implement Principal Component Analysis (PCA) (optional, given reduced features).


#### 6.1) Define a reduced set of features and target variables.

```python
# --- CRITICAL CHANGE HERE: REDUCED FEATURE SET ---
# Define a smaller, more focused set of features to manage memory.
# These features are chosen for their expected high predictive power and lower cardinality.
regression_features = [
    'Year',          # Temporal trend
    'Region',        # High-level geographical location
    'Attack Type',   # Methodology of attack
    'Group Name',    # Perpetrator
    'Weapon Type'    # Means of attack
    # 'Fatalities', 'Injuries' are part of 'Total Casualties' (target), so not included as features
]
regression_target = 'Total_Casualties'
```

```python
df_ml = df[regression_features + [regression_target]].copy()

# Identify categorical and numerical features within this reduced set
categorical_features = [col for col in regression_features if df_ml[col].dtype == 'object']
numerical_features = [col for col in regression_features if df_ml[col].dtype != 'object']

print(f"Selected Categorical features for ML: {categorical_features}")
print(f"Selected Numerical features for ML: {numerical_features}")
```

#### 6.2) Handle high-cardinality categorical features

```python
high_cardinality_cols_selected = ['Group Name'] 
top_n_thresholds_selected = {
    'Group Name': 50, 
    'Region': 15,     
    'Attack Type': 10, 
    'Weapon Type': 10 
}
```

```python
for col in high_cardinality_cols_selected:
    if col in df_ml.columns and col in categorical_features:
        top_categories = df_ml[col].value_counts().index[:top_n_thresholds_selected.get(col, 20)] # Default to 20
        df_ml[col] = np.where(df_ml[col].isin(top_categories), df_ml[col], 'Other')
        print(f"Reduced cardinality for '{col}'. New unique values: {df_ml[col].nunique()}")
```

```python
# Ensure 'Unknown' is consistently handled (grouped into 'Other' if not a top category)
if 'Group Name' in df_ml.columns and 'Unknown' in df_ml['Group Name'].unique() and 'Other' in df_ml['Group Name'].unique():
    df_ml['Group Name'] = df_ml['Group Name'].replace('Unknown', 'Other')

print("\nHigh cardinality features processed for the reduced set.")
```

#### 6.3) Apply One-Hot Encoding

```python
# Apply One-Hot Encoding to the selected categorical features
final_categorical_features_for_encoding = [col for col in categorical_features if col in df_ml.columns]
df_encoded = pd.get_dummies(df_ml, columns=final_categorical_features_for_encoding, drop_first=True)

print(f"Shape after One-Hot Encoding: {df_encoded.shape}")
```

```python
# Separate features (X) and target (y) for regression
X_reg = df_encoded.drop(columns=[regression_target])
y_reg = df_encoded[regression_target]
```

```python
# Log transformation of target
y_reg_log = np.log1p(y_reg)
print(f"Original Total Casualties skewness: {y_reg.skew():.2f}")
print(f"Log-transformed Total Casualties skewness: {y_reg_log.skew():.2f}")
```

#### 6.4) Standard Scale numerical features.

```python
scaler = StandardScaler()
numerical_features_in_X_reg = [col for col in numerical_features if col in X_reg.columns]
```

```python
# Check if there are any numerical features left to scale
if numerical_features_in_X_reg:
    X_reg_scaled = X_reg.copy() 
    X_reg_scaled[numerical_features_in_X_reg] = scaler.fit_transform(X_reg_scaled[numerical_features_in_X_reg])
    print("Numerical features scaled.")
else:
    X_reg_scaled = X_reg.copy() 
    print("No numerical features to scale in the reduced set.")

print(f"Final feature matrix shape for ML: {X_reg_scaled.shape}") 
```

### 7) Regression Models for Total Casualties

We will train and evaluate the following regression models:
- Linear Regression
- RandomForestRegressor

#### 7.1) Train-Test Split for Regression

```python
X_train_reg, X_test_reg, y_train_reg_log, y_test_reg_log = train_test_split( X_reg_scaled, y_reg_log, test_size=0.2, random_state=42)
print(f"Regression Training data shape: {X_train_reg.shape}, {y_train_reg_log.shape}")
print(f"Regression Testing data shape: {X_test_reg.shape}, {y_test_reg_log.shape}")

# Also keep the original y_test for inverse transformation and comparison
_, _, _, y_test_reg_original = train_test_split(X_reg_scaled, y_reg, test_size=0.2, random_state=42 )
```

#### 7.2) Model Training and Evaluation Loop (Regression)

```python
regression_models = {
    'Linear Regression': LinearRegression(),
    'RandomForest Regressor': RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1)
}

regression_results = {}
```

```python
for name, model in regression_models.items():
    print(f"\n--- Training {name} ---")
    try:
        model.fit(X_train_reg, y_train_reg_log)

        y_pred_log = model.predict(X_test_reg)
        # Inverse transform predictions back to original scale
        y_pred_original_scale = np.expm1(y_pred_log) # np.expm1 is inverse of np.log1p

        # Ensure predictions are non-negative
        y_pred_original_scale[y_pred_original_scale < 0] = 0

        mae = mean_absolute_error(y_test_reg_original, y_pred_original_scale)
        mse = mean_squared_error(y_test_reg_original, y_pred_original_scale)
        rmse = np.sqrt(mse)
        r2 = r2_score(y_test_reg_original, y_pred_original_scale)

        regression_results[name] = {'MAE': mae, 'MSE': mse, 'RMSE': rmse, 'R2': r2}

        print(f"  MAE: {mae:.2f}")
        print(f"  MSE: {mse:.2f}")
        print(f"  RMSE: {rmse:.2f}")
        print(f"  R2: {r2:.2f}")

    except Exception as e:
        print(f"  Error training {name}: {e}")
        regression_results[name] = {'Error': str(e)}
print("\n--- Regression Model Comparison ---")
regression_results_df = pd.DataFrame(regression_results).T
display(regression_results_df.sort_values(by='R2', ascending=False))
```

* **Regression Models:**
    * Our regression models aimed to predict the exact `Total Casualties` resulting from a terrorist attack. The results indicate that this is a particularly challenging task given the dataset and chosen features.
    * The **Decision Tree Regressor** emerged as the top performer for predicting `Total Casualties` among the regression models, achieving an **R-squared (R2) score of 0.087**. This indicates that the model explains approximately **8.7%** of the variance in the total casualties. Its **Mean Absolute Error (MAE)** was **4.84**.
    * The **RandomForest Regressor** followed with an R2 of **0.059** and an MAE of **4.74**. This means, on average, our RandomForest model's predictions for the number of casualties are off by about **4.74** individuals.
    * In contrast, simpler linear models like **Linear Regression**, **Ridge Regression**, and **Lasso Regression** performed less effectively, yielding **negative R2 scores** (e.g., -0.002 for Linear Regression). A negative R2 indicates that these models perform worse than a naive model that simply predicts the mean of the target variable for all instances.
    * This limited predictive power for exact casualty numbers is likely due to the inherent unpredictability of real-world events, the highly skewed nature of the `Total Casualties` distribution (many zeros, few extreme values), and the simplified feature set used to manage computational resources. The `log1p` transformation applied to `Total Casualties` was crucial in handling the high skewness, but the problem remains complex for precise numerical prediction.


### 8) Classification Models for Attack Severity Categories

To perform classification, we first need to define categorical severity levels based on Total Casualties.

#### 8.1) Create Severity Categories

```python
# Define severity bins and labels
# Adjust these bins based on the distribution of your 'Total Casualties'
# You can check df['Total Casualties'].describe() and histograms from EDA
bins = [-1, 0, 5, 50, df['Total_Casualties'].max()] # -1 to include 0 casualties
labels = ['No Casualties', 'Low Casualties', 'Medium Casualties', 'High Casualties']
```

```python
df_ml['Severity_Category'] = pd.cut(df_ml[regression_target], bins=bins, labels=labels, right=True)

print("\nSeverity categories created:")
print(df_ml['Severity_Category'].value_counts())
```

```python
# Prepare data for classification
# We'll use the same features as regression, but with the new target
X_clf_raw = df_ml.drop(columns=[regression_target, 'Severity_Category'])
y_clf = df_ml['Severity_Category']
```

```python
# Re-apply one-hot encoding on X_clf_raw, as pd.cut created a new column
# and we need to ensure consistency if df_ml was modified.
# It's safer to re-create the encoded features for classification.
df_encoded_clf = pd.get_dummies(df_ml.drop(columns=[regression_target, 'Severity_Category']),
                                columns=final_categorical_features_for_encoding,
                                drop_first=True)
```

```python
# Align columns - crucial if some categories were dropped in one-hot encoding for regression
# This ensures X_clf has the same columns as X_reg_scaled (minus the target)
X_clf = df_encoded_clf.reindex(columns=X_reg_scaled.columns, fill_value=0)
```

```python
# Scale numerical features for classification (using the same scaler fitted on regression data)
X_clf_scaled = X_clf.copy()
# Check if there are any numerical features left to scale
if numerical_features_in_X_reg:
    X_clf_scaled[numerical_features_in_X_reg] = scaler.transform(X_clf_scaled[numerical_features_in_X_reg])
else:
    print("No numerical features to scale in the reduced set for classification.")
print(f"Classification features shape: {X_clf_scaled.shape}")
print(f"Classification target shape: {y_clf.shape}")
```

#### 8.2) Train-Test Split for Classification

```python
X_train_clf, X_test_clf, y_train_clf_encoded, y_test_clf_encoded = train_test_split(
    X_clf_scaled, y_clf, test_size=0.2, random_state=42, stratify=y_clf # Stratify for balanced classes
)
print(f"Classification Training data shape: {X_train_clf.shape}, {y_train_clf_encoded.shape}")
print(f"Classification Testing data shape: {X_test_clf.shape}, {y_test_clf_encoded.shape}")
```

#### 8.3) Model Training and Evaluation Loop (Classification)

```python
classification_models = {
    'RandomForest Classifier': RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
}

classification_results = {}
```

```python
for name, model in classification_models.items():
    print(f"\n--- Training {name} ---")
    try:
        model.fit(X_train_clf, y_train_clf_encoded)

        y_pred_clf = model.predict(X_test_clf)

        accuracy = accuracy_score(y_test_clf_encoded, y_pred_clf)
        precision = precision_score(y_test_clf_encoded, y_pred_clf, average='weighted', zero_division=0)
        recall = recall_score(y_test_clf_encoded, y_pred_clf, average='weighted', zero_division=0)
        f1 = f1_score(y_test_clf_encoded, y_pred_clf, average='weighted', zero_division=0)
        cm = confusion_matrix(y_test_clf_encoded, y_pred_clf)
        report = classification_report(y_test_clf_encoded, y_pred_clf, zero_division=0)

        classification_results[name] = {
            'Accuracy': accuracy,
            'Precision': precision,
            'Recall': recall,
            'F1-Score': f1,
            'Confusion Matrix': cm,
            'Classification Report': report
        }

        print(f"  Accuracy: {accuracy:.2f}")
        print(f"  Precision (weighted): {precision:.2f}")
        print(f"  Recall (weighted): {recall:.2f}")
        print(f"  F1-Score (weighted): {f1:.2f}")
        print("\n  Confusion Matrix:\n", cm)
        print("\n  Classification Report:\n", report)

    except Exception as e:
        print(f"  Error training {name}: {e}")
        classification_results[name] = {'Error': str(e)}

print("\n--- Classification Model Comparison ---")
classification_results_df = pd.DataFrame({k: {m: v for m, v in val.items() if m not in ['Confusion Matrix', 'Classification Report']} for k, val in classification_results.items()}).T
display(classification_results_df.sort_values(by='Accuracy', ascending=False))
```
